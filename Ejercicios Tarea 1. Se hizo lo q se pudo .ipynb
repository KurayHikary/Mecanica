```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
#Ejercicio 2.9 C. Trayectoria en 3D del carro
fig=plt.figure()
ax=Axes3D(fig)
t = np.linspace(0,20,20*np.pi)
r = np.linspace (0, 50, len(t))
z = np.linspace (0, 50, len(t))
ax.plot(r*np.sin(t), r*np.cos(t), z, color='darkgreen')
plt.title("Trayectoria de la masa m dentro del cono")
plt.xlabel("X", size = 16,)
plt.ylabel("Y", size = 16)
plt.show()
```

```python
#Ejercicio 2.17 C Evaluacion numerica para la trayectoria de la masa 
#Elegimos valores numericos para los coeficientes de friccion 
fig=plt.figure()
ax=Axes3D(fig)
t=np.linspace(0,3,300)
m=1
g=10 
f_1=0.4
f_2=0.6
angulo=np.pi/4
R=-g*f_2
i=((m*g*(np.sin(angulo)-f_1*np.cos(angulo)))-(R*(np.cos(angulo)+f_1*np.sin(angulo)))/(np.cos(angulo)+np.tan(angulo)*np.sin(angulo)))
j=((R*(np.cos(angulo)+f_1*np.sin(angulo)))-(m*g*(np.sin(angulo)-f_1*np.cos(angulo))))/(np.cos(angulo)+np.tan(angulo)*np.sin(angulo))
r_x=f_1*(i+R)*t**2
r_y=f_1*np.tan(angulo)*j*t**2
ax.plot(r_x,r_y,t)
plt.title("Trayectoria de la masa en 3D Evaluacion numerica")
plt.show()

```

```python
#Ejercicio 2.29 C. Trayectoria en 3D del carro
t = np.linspace(0,20,500)
def trayectoria_carro(t):
    x=t*np.cos(t)
    y=t*np.sin(t)
    z=1
    return x,y,z
ax.plot(trayectoria_carro(t)[0],trayectoria_carro(t)[1],trayectoria_carro(t)[2], color='rebeccapurple')
plt.title("Trayectoria Del Carro Manteniendo Velocidad Angular Constante")
plt.xlabel("X", size = 16,)
plt.ylabel("Y", size = 16)
plt.show()
```

```python
#Ejercicio 1.  Propuesto en Clase (Cuerda)
#Caso 1 lambda(r)=lambda_0
#Consideramos theta= pi/2, lambda=1 y w=stqr(2) rad/s
r=np.linspace(0,10,20)
g=10 #gravedad
T=-r**2+g*r
plt.plot(r,T)
plt.axvline(g/2, color="green", linestyle="--")
plt.text(5, 0, 'g/2', fontsize=12, color='darkgreen')
plt.axhline((g**2)/4, color="red", linestyle="dashed")
plt.text(0, 23, '(lambda*g^2)/4', fontsize=12, color='red')
plt.title('Caso 1 lambda(r)=lambda_0')
plt.xlabel('r')
plt.ylabel('T(r)')
plt.show()
```

```python
#Caso 2 lambda(r)=C*r
#Consideramos theta= pi/2, C=1, y w=stqr(3) rad/s

r=np.linspace(0,5,50)
g=10 #gravedad
T=(-r**3)+((g/2)*(r**2))
plt.plot(r,T)
plt.axvline(g/2, color="green", linestyle="--")
plt.text(4.6, 0, 'g/2', fontsize=12, color='darkgreen')
plt.axvline(g/3, color="red", linestyle="--")
plt.text(2.9, 0, 'g/3', fontsize=12, color='red')
plt.axhline((g**3)/54, color="purple", linestyle="dashed")
plt.text(0, 17, 'C*(g^3)/54', fontsize=12, color='purple')
plt.title('Caso 2 lambda(r)=C*r')
plt.xlabel('r')
plt.ylabel('T(r)')
plt.show()
```
